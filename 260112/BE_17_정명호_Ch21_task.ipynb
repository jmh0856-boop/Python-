[문제 6] 0부터 시작하여 n까지의 숫자 중에서 홀수만 생성하는 제너레이터 odd_numbers를 작성하세요.

```python
def odd_numbers(n):
    for i in range(n+1):
        if i % 2 != 0:
            yield i

for num in odd_numbers(20):
    print(num, end=" ")
```

```python
numbers = tuple(map(int, input().split()))

def min_max(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    print(f"최소값: {min_val}, 최대값: {max_val}")

min_max(numbers)
```

[문제 10] 팩토리얼을 계산하는 재귀 함수 factorial을 작성하세요.

```python
def factorial(n):
    if n ==1:
        return 1
    return n * factorial(n - 1)
print(factorial(5))
```

[문제 11] 문자열 리스트를 받아, 각 문자열의 길이를 출력하는 함수 print_lengths를 작성하세요.

```python
def print_lengths(strings):
    for string in strings:
        print(len(string))
print_lengths(["hello", "world!"])
```

[문제 12] 두 문자열을 받아서 긴 문자열을 반환하는 함수 longer_string을 작성하세요. 만약 두 문자열의 길이가 같다면, 첫 번째 문자열을 반환하세요.

```python
def longer_string(str1, str2):
    if len(str1) >= len(str2):
        return str1
    else:
        return str2
longer_string("hello", "world!")
```
