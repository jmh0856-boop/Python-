## 모험가들의 레벨업 도전

### [배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다. 

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

<br>

### [동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 

마법사의 지능이 50 이상인지 확인하세요.

```python
Mage_stats = {'STR': 10, 'DEX': 20, 'INT':80, 'CON': 50}
Warrior_stats = {'STR': 80, 'DEX': 30, 'INT':20, 'CON': 100}
Rogue_stats = {'STR': 40, 'DEX': 80, 'INT':30, 'CON': 60}
Healer_stats = {'STR': 20, 'DEX': 40, 'INT':70, 'CON': 70}
Use_Magic = Mage_stats['INT'] >= 50
print("마법사의 지능이 50 이상:", Use_Magic)
```

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다. 

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
Mage_stats = {'STR': 10, 'DEX': 20, 'INT':80, 'CON': 50}
Warrior_stats = {'STR': 80, 'DEX': 30, 'INT':20, 'CON': 100}
Rogue_stats = {'STR': 40, 'DEX': 80, 'INT':30, 'CON': 60}
Healer_stats = {'STR': 20, 'DEX': 40, 'INT':70, 'CON': 70}
Warrior_Attack = Warrior_stats['STR'] >= 60 and Warrior_stats['CON'] >=80
print("전사의 힘이 60이상, 체력이 80 이상:", Warrior_Attack)
```

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다. 

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
Mage_stats = {'STR': 10, 'DEX': 20, 'INT':80, 'CON': 50}
Warrior_stats = {'STR': 80, 'DEX': 30, 'INT':20, 'CON': 100}
Rogue_stats = {'STR': 40, 'DEX': 80, 'INT':30, 'CON': 60}
Healer_stats = {'STR': 20, 'DEX': 40, 'INT':70, 'CON': 70}
Rogue_Attack = Rouge_stats['DEX'] >= 70 or ['STR'] >= 50
print("도적의 민첩이 70 이상이거나 힘이 50 이상:", Rogue_Attack)
```

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다. 

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
Mage_stats = {'STR': 10, 'DEX': 20, 'INT':80, 'CON': 50}
Warrior_stats = {'STR': 80, 'DEX': 30, 'INT':20, 'CON': 100}
Rogue_stats = {'STR': 40, 'DEX': 80, 'INT':30, 'CON': 60}
Healer_stats = {'STR': 20, 'DEX': 40, 'INT':70, 'CON': 70}
Use_healing_Magic = Healer_stats['INT'] >= 60 and Healer_stats['CON'] >= 50
print("힐러의 지능이 60 이상이고, 체력이 50 이상:", Use_healing_Magic)
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다. 


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
Exp = {'Mage': 120, 'Warrior': 150, 'Rogue': 110, 'Healer': 90}
Mage_Level_Up = Exp['Mage'] >= 100
Warrior_Level_Up = Exp['Warrior'] >= 100
Rogue_Level_Up = Exp['Rogue'] >= 100
Healer_Level_Up = Exp['Healer'] >= 100 
print("마법사:",Mage_Level_Up, "전사:",Warrior_Level_Up, "도적:",Rogue_Level_Up, "힐러:",Healer_Level_Up)
```
