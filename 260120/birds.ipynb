```python
class Birds:
    def __init__(self, name, tweet, fly, distance):
        self.name = name
        self.tweet = tweet
        self.fly = fly
        self.distance = distance

    def display(self):
        return f"{self.name} 출발!!!\n{self.tweet}\n{self.fly} 날았습니다\n{self.distance} m"

birds_dict = {
    '앵무새': Birds(name='앵무새', tweet='까악', fly='날개를 힘차게', distance=2),
    '참새': Birds(name='참새', tweet='짹짹', fly='날개를 빠르게', distance=8),
    '비둘기': Birds(name='비둘기', tweet='구구', fly='날개를 부드럽게', distance=12),
    '닭': Birds(name='닭', tweet='꼬끼오', fly='날개를 퍼덕이며', distance=3),
    '러버덕': Birds(name='러버덕', tweet='빽빽빽', fly='날지 못함', distance=0)
}

user_input_name = input("앵무새,참새,비둘기,닭,러버덕 중 선택해 주세요")

if user_input_name in birds_dict:
    select_bird = birds_dict[user_input_name]
    print(select_bird.display())
else:
    print("해당하는 새 이름은 없습니다.")
```
